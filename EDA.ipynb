```python
import pandas as pd
import numpy as np
import gc
```

```python
train = pd.read_csv("train.csv")
```

```python
train.head()
```

```python
test = pd.read_csv("test.csv")
```

```python
test.head()
```

```python
def get_cate(df):
    def get_last(x):
        tmp = x.split('_')
        return int(tmp[1])
        
    for col in ['A','B','C','D','E','F']:
        df[f'category{col}'] = df[f'category{col}'].map(get_last)
    
    df['unit'] = df['unit'].map(get_last)
    return df
```

```python
train = get_cate(train)
test = get_cate(test)
```

```python
train.info()
```

```python
columns = ['categoryA', 'categoryB', 'categoryC', 'categoryD', 'categoryE',
       'categoryF', 'featureA', 'featureB', 'featureC', 'featureD', 'featureE',
       'featureF', 'featureG', 'featureH', 'featureI', 'compositionA',
       'compositionB', 'compositionC', 'compositionD', 'compositionE',
       'compositionF', 'compositionG', 'compositionH', 'compositionI',
       'compositionJ', 'unit']
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import KFold
NFOLDS = 5
folds = KFold(n_splits=NFOLDS)
X_train = train[columns]
y_train = train['result']
splits = folds.split(train, y_train)
X_test = test[columns]
```

```python
y_preds = np.zeros(X_test.shape[0])
y_oof = np.zeros(X_train.shape[0])
cat_features = [0,1,2,3,4,5,25]

for fold_n, (train_index, valid_index) in enumerate(splits):
    X_tr, X_val = train[columns].iloc[train_index], train[columns].iloc[valid_index]
    y_tr, y_val = train['result'].iloc[train_index], train['result'].iloc[valid_index]    
    
    clf= CatBoostRegressor(n_estimators = 20000,
                           depth= 8,
                           eta= 0.03,
                           loss_function = 'RMSE',
                           eval_metric = 'RMSE',
                           random_strength= 1.5,
                           reg_lambda= 6,
                           od_type= 'Iter',
                           border_count= 128,
                           bootstrap_type = "Bayesian")

    clf.fit(X_tr, y_tr,eval_set=(X_val, y_val),early_stopping_rounds=100,cat_features=cat_features)
    y_pred_valid = clf.predict(X_val)
    y_oof[valid_index] = y_pred_valid
#     print(f"Fold {fold_n + 1} | AUC: {roc_auc_score(y_val, y_pred_valid)}")   
#     score += roc_auc_score(y_val, y_pred_valid) / NFOLDS
    y_preds += clf.predict(X_test)/ NFOLDS    
    del X_tr, X_val, y_tr, y_val
    gc.collect()       
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub['result'] = y_preds
```

```python
sub.to_csv('sub_test.csv',index=None)
```
