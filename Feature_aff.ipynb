```python
category = ['categoryA', 'categoryB', 'categoryC', 'categoryD', 'categoryE','categoryF','unit']
numerical = ['featureA', 'featureB', 'featureC', 'featureD', 'featureE',
             'featureF', 'featureG', 'featureH', 'featureI', 'compositionA',
             'compositionB', 'compositionC', 'compositionD', 'compositionE',
             'compositionF', 'compositionG', 'compositionH', 'compositionI','compositionJ']
```

```python
def feature_agg(df,category,numerical):
    for C in category:
        for N in numerical:
            
            df[C+'_'+N+'_median'] = df.groupby(C)[N].transform('median')
            df[C+'_'+N+'_mean'] =df.groupby(C)[N].transform('mean')
            df[C+'_'+N+'_min'] =df.groupby(C)[N].transform('min')
            df[C+'_'+N+'_max'] =df.groupby(C)[N].transform('max')
            df[C+'_'+N+'_std'] =df.groupby(C)[N].transform('std')
            df[C+'_'+N+'_var'] =df.groupby(C)[N].transform('var')
            df[C+'_'+N+'_count'] =df.groupby(C)[N].transform('count')

            df[N+f'+Median({C})'] = df[N] + df[C+'_'+N+'_median']
            df[N+f'-Median({C})'] = df[N] - df[C+'_'+N+'_median']
            df[N+f'*Median({C})'] = df[N] * df[C+'_'+N+'_median']
            df[N+f'/Median({C})'] = df[N] / df[C+'_'+N+'_median']

            df[N+f'+Mean({C})'] = df[N] + df[C+'_'+N+'_mean']
            df[N+f'-Mean({C})'] = df[N] - df[C+'_'+N+'_mean']
            df[N+f'*Mean({C})'] = df[N] * df[C+'_'+N+'_mean']
            df[N+f'/Mean({C})'] = df[N] / df[C+'_'+N+'_mean']
    return df
from sklearn import preprocessing
def add_compound_features(df1,df2,features_to_compound):
    for f1 in features_to_compound:
        for f2 in features_to_compound:
            if f1 != f2:
                df1[f1+'_'+f2] = df1[f1].astype(str) + '_' + df1[f2].astype(str)
                df2[f1+'_'+f2] = df2[f1].astype(str) + '_' + df2[f2].astype(str)
                le = preprocessing.LabelEncoder()
                le.fit(df1[f1+'_'+f2])
                df2[f1+'_'+f2] = df2[f1+'_'+f2].map(lambda s: '<unknown>' if s not in le.classes_ else s)
                le.classes_ = np.append(le.classes_, '<unknown>')
                df1[f1+'_'+f2] = le.transform(df1[f1+'_'+f2]) 
                df2[f1+'_'+f2] = le.transform(df2[f1+'_'+f2]) 
    return df1,df2

def label_encoder(df1,df2,features):
    for f in features:
        le = preprocessing.LabelEncoder()
        le.fit(df1[f])
        df2[f] = df2[f].map(lambda s: '<unknown>' if s not in le.classes_ else s)
        le.classes_ = np.append(le.classes_, '<unknown>')
        df1[f] = le.transform(df1[f]) 
        df2[f] = le.transform(df2[f]) 
    return df1,df2
```

```python
train = feature_agg(train,category,numerical)
test = feature_agg(test,category,numerical)
print(test.shape)
train,test = add_compound_features(train,test,category)
print(test.shape)
train,test = label_encoder(train,test,category)
```

```python
columns = list(test.columns)
columns.remove('id')
# columns.remove('result')
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import KFold
NFOLDS = 5
folds = KFold(n_splits=NFOLDS)
X_train = train[columns]
y_train = train['result']
splits = folds.split(train, y_train)
X_test = test[columns]
```

```python
y_preds = np.zeros(X_test.shape[0])
y_oof = np.zeros(X_train.shape[0])
cat_features = [0,1,2,3,4,5,25]# + [2021+i for i in range(42)]

for fold_n, (train_index, valid_index) in enumerate(splits):
    X_tr, X_val = train[columns].iloc[train_index], train[columns].iloc[valid_index]
    y_tr, y_val = train['result'].iloc[train_index], train['result'].iloc[valid_index]    
    
    clf= CatBoostRegressor(n_estimators = 20000,
                           depth= 8,
                           eta= 0.03,
                           loss_function = 'RMSE',
                           eval_metric = 'RMSE',
                           random_strength= 1.5,
                           reg_lambda= 50,
                           od_type= 'Iter',
                           border_count= 128,
                           bootstrap_type = "Bayesian"
                          )

    clf.fit(X_tr, y_tr,eval_set=(X_val, y_val),early_stopping_rounds=100,cat_features=cat_features,verbose_eval=200)
#     clf.fit(X_tr, y_tr,eval_set=(X_val, y_val),early_stopping_rounds=200,verbose_eval=200)
    y_pred_valid = clf.predict(X_val)
    y_oof[valid_index] = y_pred_valid
#     print(f"Fold {fold_n + 1} | AUC: {roc_auc_score(y_val, y_pred_valid)}")   
#     score += roc_auc_score(y_val, y_pred_valid) / NFOLDS
    y_preds += clf.predict(X_test)/ NFOLDS    
    del X_tr, X_val, y_tr, y_val
    gc.collect()       
    break
```
